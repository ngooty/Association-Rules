## Association Rule Mining - Market Basket Analysis

```python
# Trick/pre requisite
# for association rule mining - your dataset should be a transcation dataset/ list of lists
# normally we get a table,excelsheet, csv. we have to convert this dataset to a transcation set for assoc rule mining. 
```

```python
#couple of library which are popular 
# mlxtend library
# apyori library - preferable
```

```python
import pandas as pd
import numpy as np
from apyori import apriori
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
store=pd.read_csv('store_data.csv',header=None)
```

```python
store.sample(6)
```

```python


store.head()
#store.values[0,2]
```

```python
#convert df to list of lists
records=[]
for i in range(0,7501):
    records.append([str(store.values[i,j]) for j in range(0,20)])
```

```python
records[1]
```

```python
association_rules=apriori(records,min_support=0.004,min_confidence=0.2,min_lift=3,min_length=2)
rules=list(association_rules)
```

```python
len(rules)
```

```python
rules[1:10]
```

```python
for items in rules:
    pair=items[0]
    paired_item=[x for x in pair]
    print('Rule: '+paired_item[0] + "=>"+paired_item[1]+'  support: '+ str(round(items[1],3))
         +'  Confidence: '+str(round(items[2][0][2],3))+'  Lift: '+str(round(items[2][0][3],3)))
    print('********************************')
    
    
```
